### Pick a time in which you want to see the ferry data
This will slice the entire data set to the selected time

```python
data_A2016=ferryQC.sel(time=slice('2016-11-01T18:10:00', '2016-11-01T18:30:00'))
```

```python
data_A2016.Horizontal_speed
```

### Plot the horizontal speed:

```python
%matplotlib inline
plt.figure(figsize=(14,4))
data_A2016.Horizontal_speed.plot(x='time', y='depth')
plt.gca().invert_yaxis()
data_A2016.bottom_tracking_depth_beam_1.plot()
```

### We can select a depth and then plot time series

This does not look very well

```python
data_depth=data_A2016.sel(depth=14.05, method='nearest')
```

```python
plt.figure(figsize=(14,4))
data_depth.Horizontal_speed.plot()
```

### Plot data in space

```python
plt.figure(figsize=(14,4))
plt.scatter(data_A2016.longitude,data_A2016.latitude,c=data_depth.Horizontal_speed)
```
